```python
import pandas
from tqdm import tqdm
```

```python
crime_filepath = "CrimeWatch_Maps_Past_90-Days.csv"
crime_data = pandas.read_csv(crime_filepath)
print(len(crime_data))
crime_data['lat'] = 0
crime_data['lng'] = 0
crime_data.head()
```

```python
import googlemaps
from datetime import datetime

gmaps = googlemaps.Client(key=open("secrets/google_maps_key.txt").read())
```

```python
for index in tqdm(range(len(crime_data))):
    address = crime_data.loc[index, "Location"].replace("\n", ", ")
    geocode_result = gmaps.geocode(address)
    if len(geocode_result):
        lat_lng = geocode_result[0]['geometry']['location']
        crime_data.loc[index, "lat"] = lat_lng['lat']
        crime_data.loc[index, "lng"] = lat_lng['lng']
crime_data.head()
```

```python
crime_data.to_csv('CrimeWatch_Maps_Past_90-Days_geocoded.csv')
```
